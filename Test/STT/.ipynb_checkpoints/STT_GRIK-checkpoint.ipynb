```python
import wave
import json
from vosk import Model, KaldiRecognizer

# Vosk 한글 모델 로드
model = Model("vosk-model-small-ko-0.22")

# WAV 파일 로드
wf = wave.open("/content/drive/MyDrive/IAP/STT/인공지능플랫폼_16-02.wav", "rb")

# 인식기 생성
rec = KaldiRecognizer(model, wf.getframerate())

# 음성 인식 실행
while True:
    data = wf.readframes(4000)
    if len(data) == 0:
        break
    if rec.AcceptWaveform(data):
        result = rec.Result()
        print(json.loads(result))

# 마지막 결과 출력
print(json.loads(rec.FinalResult()))
```

## Whisper로 텍스트로 추출하고 파일 저장하기

```python
import whisper

# Using Tiny Model
model = whisper.load_model("tiny", device="cuda")

# 음성 파일로부터 텍스트 추출
result = model.transcribe("/content/test.mp3", language='en')

# 결과 출력
print(result["text"])

# 결과를 txt로 저장
w = open("test.txt", 'w')
w.write(result["text"])
w.close()
```
