```python
import cv2
import numpy as np
#from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
#!pwd
```

```python
def imshow(image):
  plt.imshow(image)
  plt.axis('off')
  plt.show()
```

```python
printed = cv2.imread('Compare/new_compare_original.jpg')
handwritten = cv2.imread('Compare/new_compare_handwrite.jpg')

pr = cv2.cvtColor(printed, cv2.COLOR_BGR2RGB)
hw = cv2.cvtColor(handwritten, cv2.COLOR_BGR2RGB)

imshow(pr)
imshow(hw)
```

```python
print(pr.shape)
print(hw.shape)
```

```python
h, w, c = pr.shape # h: height, w: weight, c: channel

diff = np.abs(hw.astype(np.int32) - pr.astype(np.int32))
diff_cv2 = cv2.cvtColor(diff.astype(np.uint8), cv2.COLOR_RGB2BGR)
diff_cv2 = cv2.bitwise_not(diff_cv2)
diff = (cv2.cvtColor(diff_cv2, cv2.COLOR_BGR2RGB)).astype(np.int32)

mask = (diff > 225).all(axis=2)
hw[mask] = (255, 255, 255)
imshow(hw)
```
